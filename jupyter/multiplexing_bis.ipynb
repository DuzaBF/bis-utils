# Multiplexing 4 electrode array for bioimpedance spectroscopy of the limbs
Using one-layer and two-layer half-space model with am assumption of the quasi static case.

# One-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{1}{\sqrt{r^2 + z^2}},
$$
where 

$I$ - injected current;

$\sigma$ - conductivity of the half-space material;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{r}{(r^2 + z^2)^{3/2}}
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{z}{(r^2 + z^2)^{3/2}}
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma} \cdot \frac{1}{|r|}
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma} \cdot \frac{1}{r^2}
$$
$$
E_z = 0
$$

# Two-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{\sqrt{r^2 + z^2}} + 
    \sum^{\infty}_{n=1}K^{n}_{12}
    \left(
        \frac{1}{\sqrt{r^2 + (2nd_1-z)^2}} + 
        \frac{1}{\sqrt{r^2 + (2nd_1+z)^2}} 
    \right) 
\right],
$$
where 

$I$ - injected current;

$K_{12} = \frac{\sigma_1 - \sigma_2}{\sigma_1 + \sigma_2}$ - reflection coefficient; 

$\sigma_1$ - conductivity of the first layer;

$\sigma_2$ - conductivity of the second layer;

$d_1$ - thickness of the first layer;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{r}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{rK^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{z}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{-(2nd_1 - z)K^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{(2nd_1 + z)K^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{|r|} + 
    2\sum^{\infty}_{n=1}\frac{K^{n}_{12}}{\sqrt{r^2 + (2nd_1)^2}} 
\right],
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{r^2} + 
    2\sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1)^2)^{3/2}} 
    \right) 
\right]
$$
$$
E_z = 0
$$

# Tetrapolar electrode array
The 4 electrodes make an tetrapolar electrode array. The electrodes are numbered 1-4. The coordinates of the electrodes are denoted as $x_1$, $x_2$, $x_3$, $x_4$.The distance between electrodes 1 and 4 is $L$, between electrodes 2 and 3 is $s$. 
In symmetrical tetrapolar electrode array the current is injected through the electrodes denoted A and B and voltage is measured between electrodes denoted M and N. On the surface the potential will depend only on the distance from the electrode to the measurement point.
For simplicity:
$$
q=\frac{I}{2 \pi \sigma_1}
$$
The potentials at points M is determined as a sum of potentials from electrode A with current $I$ and electrode B with current $-I$. The subscripts denote position of the corresponding electrode.
$$
V_M = V(x_M - x_A) + V(x_M - x_B) = q \cdot 
\left[
        \frac{1}{|x_M - x_A|} - \frac{1}{|x_M - x_B|} + 
        2\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} 
        \right)
\right]
$$
$$
V_N = V(x_N - x_A) + V(x_N - x_B) = q \cdot 
\left[
        \frac{1}{|x_N - x_A|} - \frac{1}{|x_N - x_B|} +
        2\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}} 
        \right)
\right]
$$

Voltage between electrodes:
$$
\begin{split}
U_{MN} = V_M - V_N  = q \cdot
&\left[
        \frac{1}{|x_M - x_A|} - \frac{1}{|x_M - x_B|} -  \frac{1}{|x_N - x_A|} + \frac{1}{|x_N - x_B|} \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}}
        \right)
\right]
\end{split}
$$
Usually the current is injected through electrodes 1 and 4 and voltage is measured between electrodes 2 and 3.
$$
\begin{split}
U_{MN} = q \cdot
&\left[
        \frac{4}{L-s} - \frac{4}{L+s}
        + 4\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(\frac{L-s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}}
        \right)
\right]
\end{split}
$$

Measured impedance:
$$
Z_{MN} = \frac{U_{MN}}{I} = \frac{1}{2 \pi \sigma_1} \cdot 
\left[
        \frac{4}{L-s} - \frac{4}{L+s} + 
        4\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(\frac{L-s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} 
        \right)
\right]
$$

By changing which pairs of the electrodes are current injecting and which are voltage measuring different impedance will be measured. Due to symmetry and reciprocity only 3 arrangements are distinct:
1. 1, 4 - current, 2, 3 - measuring;
2. 3, 4 - current, 1, 2 - measuring;
3. 2, 4 - current, 1, 3 - measuring.

The measured impedances in case 2 are determined by:
$$
\begin{split}
Z_{12} = \frac{1}{2 \pi \sigma_1} \cdot
&\left[
        \frac{1}{|x_1 - x_4|} - \frac{1}{|x_1 - x_3|} - \frac{1}{|x_2 - x_4|} + \frac{1}{|x_2 - x_3|} \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_1 - x_4)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_1 - x_3)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_2 - x_4)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_2 - x_3)^2 + (2nd_1)^2}} 
        \right)
\right] = \\
= 
&\left[
        \frac{1}{L} - \frac{1}{\frac{L+s}{2}} - \frac{1}{\frac{L+s}{2}} + \frac{1}{s} \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(L)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(s)^2 + (2nd_1)^2}} 
        \right)
\right]
\end{split}
$$
$$
Z_{12} = \frac{1}{2 \pi \sigma_1} \cdot
\left[
        \frac{1}{L} + \frac{1}{s} - \frac{4}{L+s}
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}.\left(
            \frac{1}{\sqrt{L^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{s^2 + (2nd_1)^2}} - 
            \frac{2}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}}
        \right)
\right]
$$
The measured impedances in case 3 are determined by:
$$
\begin{split}
Z_{13} = \frac{1}{2 \pi \sigma_1} \cdot
&\left[
        \frac{1}{|x_1 - x_4|} - \frac{1}{|x_1 - x_2|} - \frac{1}{|x_3 - x_4|} + \frac{1}{|x_3 - x_2|} \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_1 - x_4)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_1 - x_2)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_3 - x_4)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_3 - x_2)^2 + (2nd_1)^2}} 
        \right)
\right] = \\
= 
&\left[
        \frac{1}{L} - \frac{1}{\frac{L-s}{2}} - \frac{1}{\frac{L-s}{2}} + \frac{1}{s}  \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(L)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(s)^2 + (2nd_1)^2}} 
        \right)
\right]
\end{split}
$$
$$
Z_{13} = \frac{1}{2 \pi \sigma_1} \cdot
\left[
        \frac{1}{L} + \frac{1}{s} - \frac{4}{L-s}
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}.\left(
            \frac{1}{\sqrt{L^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{s^2 + (2nd_1)^2}} -
            \frac{2}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}}
        \right)
\right]
$$

For simplicity, the geometrical factor of the tetrapolar electrode array on the one-layer model is defined as:
$$
G_1(x_A, x_M, x_N, x_B) = \left|\frac{1}{|x_M - x_A|} - \frac{1}{|x_M - x_B|} -  \frac{1}{|x_N - x_A|} + \frac{1}{|x_N - x_B|}\right|
$$
The geometrical factor correction for the two-layer model. The $G_2$ factor is also a function of integer $n$ and teh first layer depth $d_1$:
$$
\begin{split}
G_2(x_A, x_M, x_N, x_B, n) = 
        &\left| \phantom{+}\frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} \right. \\
        &-\frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}}           \\
        &- \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} \\
        &\left.+\frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}} \right|
\end{split}
$$

As the impedance is positive and doesn't depend on with electrode is current source and which is current sink the geometrical factors are taken as absolute values. 

The impedance is than equal:
$$
Z(x_A, x_M, x_N, x_B) = \frac{1}{2 \pi \sigma_1}\left[G_1(x_A, x_M, x_N, x_B) + 2\sum^{\infty}_{n=1}
        K^{n}_{12}
        G_2(x_A, x_M, x_N, x_B, n)
\right]
$$

# Least squares estimation from 3 measurements

The impedance is a function of 4 variables: position of electrodes $x_A$ $x_M$, $x_N$, $x_B$.
$$
\begin{split}
Z(x_A, x_M, x_N, x_B) = \frac{1}{2  \pi  \sigma_1} \cdot
&\left[
        \frac{1}{|x_M - x_A|} - \frac{1}{|x_M - x_B|} -  \frac{1}{|x_N - x_A|} + \frac{1}{|x_N - x_B|} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}}
        \right)
\right]
\end{split}
$$
For simplicity a given electrodes arrangement can be denoted as vector:
$$
\textbf{X} = 
\begin{bmatrix}
x_A \\
x_M \\
x_N \\
x_B 
\end{bmatrix}
$$

The $\sigma_1$, $\sigma_2$, $d_1$ are the unknown properties of the studied object. To determine properties from the impedance measurements the Least Squares method can be used. The idea is to find such $\sigma_1$, $\sigma_2$, $d_1$ which minimize the square error from the measurements $Z_1..Z_n$ for the electrode arrangements $\textbf{X}_1..\textbf{X}_n$.
$$
D = \sum_{i=1}^{n}\left(Z_i - Z(\textbf{X}_i)\right)^2 \rightarrow min
$$

The residual function is:
$$
r(\sigma_1, \sigma_2, d_1) = 
\begin{bmatrix}
    Z_1 - Z(\textbf{X}_1) \\
    Z_2 - Z(\textbf{X}_2) \\
    Z_3 - Z(\textbf{X}_3)
\end{bmatrix}
$$

To find $\sigma_1$, $\sigma_2$, $d_1$ we find the partial derivatives of $D$ and find their zeros:
$$
\frac{\partial D}{\partial \sigma_1} = \sum_{i=1}^{n}
2(Z_i - Z(\textbf{X}_i))
\left. \frac{\partial Z(\textbf{X})}{\partial \sigma_1} \right\rvert_{\textbf{X}_i} = 0
$$

$$
\frac{\partial D}{\partial \sigma_2} = \sum_{i=1}^{n}
2(Z_i - Z(\textbf{X}_i))
\left. \frac{\partial Z(\textbf{X})}{\partial \sigma_2} \right\rvert_{\textbf{X}_i} = 0
$$

$$
\frac{\partial D}{\partial d_1} = \sum_{i=1}^{n}
2(Z_i - Z(\textbf{X}_i))
\left. \frac{\partial Z(\textbf{X})}{\partial d_1} \right\rvert_{\textbf{X}_i} = 0
$$


The Jacobian of the residual function is:
$$
J_r(\sigma_1, \sigma_2, d_1) = 
\begin{bmatrix}
\frac{\partial Z(\textbf{X}_1)}{\partial \sigma_1}
\frac{\partial Z(\textbf{X}_1)}{\partial \sigma_2}
\frac{\partial Z(\textbf{X}_1)}{\partial d_1}
\\[10pt]
\frac{\partial Z(\textbf{X}_2)}{\partial \sigma_1} 
\frac{\partial Z(\textbf{X}_2)}{\partial \sigma_2} 
\frac{\partial Z(\textbf{X}_2)}{\partial d_1} 
\\[10pt]
\frac{\partial Z(\textbf{X}_3)}{\partial \sigma_1} 
\frac{\partial Z(\textbf{X}_3)}{\partial \sigma_2} 
\frac{\partial Z(\textbf{X}_3)}{\partial d_1}  
\end{bmatrix} 
$$ 


The partial derivative with respect to $\sigma_1$ is equal:
$$
\begin{split}
\frac{\partial Z(\textbf{X})}{\partial \sigma_1} &= 
\frac{\partial }{\partial \sigma_1}
\left[
\frac{1}{2 \pi \sigma_1} \cdot
\left(
G_1(\textbf{X}) + 2\sum^{\infty}_{n=1}
        K^{n}_{12}G_2(\textbf{X}, n)
\right)
\right]  \\
&=
- \frac{1}{2 \pi \sigma_1^2}
\left[
G_1(\textbf{X}) + 2\sum^{\infty}_{n=1}
        K^{n}_{12}G_2(\textbf{X}, n)
\right]  \\
&\phantom{=}+\frac{1}{2 \pi \sigma_1}
\left[
2\sum^{\infty}_{n=1}
        nK^{n-1}_{12}\frac{\sigma_2}{(\sigma_1+\sigma_2)^2}
        G_2(\textbf{X}, n)
\right]  \\
&=- \frac{1}{2 \pi \sigma_1^2}
\left[
G_1(\textbf{X})+2\sum^{\infty}_{n=1}
 K^{n}_{12}G_2(\textbf{X}, n)\left(1 - \frac{2\sigma_1 \sigma_2}{\sigma_1^2 - \sigma_2^2}n\right)
\right]
\end{split}
$$
The partial derivative with respect to $\sigma_2$ is equal:
$$
\begin{split}
\frac{\partial Z(\textbf{X})}{\partial \sigma_2} &= 
\frac{\partial }{\partial \sigma_2}
\left[
\frac{1}{2 \pi \sigma_1} \cdot
\left(
G_1(\textbf{X}) + 2\sum^{\infty}_{n=1}
K^{n}_{12}G_2(\textbf{X}, n)
\right)
\right]  \\
&= \frac{1}{2 \pi \sigma_1} \cdot
\frac{\partial}{\partial \sigma_2}2\sum^{\infty}_{n=1}
K^{n}_{12}G_2(\textbf{X}, n)  \\
&= -\frac{1}{2 \pi \sigma_1} \cdot
2\sum^{\infty}_{n=1}
nK^{n}_{12}G_2(\textbf{X}, n)\frac{2\sigma_1}{\sigma_1^2 - \sigma_2^2} \\
&=
-\frac{2}{\pi(\sigma_1^2 - \sigma_2^2)}
\sum^{\infty}_{n=1}
nK^{n}_{12}G_2(\textbf{X}, n)
\end{split}
$$
The partial derivative with respect to $d_1$ is equal:
$$
\begin{split}
\frac{\partial Z(\textbf{X})}{\partial d_1} &= 
\frac{\partial }{\partial d_1}
\left[
\frac{1}{2 \pi \sigma_1} \cdot
\left(
G_1(\textbf{X}) + 2\sum^{\infty}_{n=1}
K^{n}_{12}G_2(\textbf{X}, n)
\right)
\right]  \\
&= \frac{1}{2 \pi \sigma_1} \cdot
2\sum^{\infty}_{n=1}
K^{n}_{12}\frac{\partial G_2(\textbf{X}, n)}{\partial d_1}
\end{split}
$$

$$
\begin{split}
\frac{\partial G_2(x_A, x_M, x_N, x_B, n)}{\partial d_1}
 =
    &- \frac{4 d_1 n^2}{((x_M - x_A)^2 + (2nd_1)^2)^{3/2}}  \\
    &+ \frac{4 d_1 n^2}{((x_M - x_B)^2 + (2nd_1)^2)^{3/2}}  \\
    &+ \frac{4 d_1 n^2}{((x_N - x_A)^2 + (2nd_1)^2)^{3/2}}  \\
    &- \frac{4 d_1 n^2}{((x_N - x_B)^2 + (2nd_1)^2)^{3/2}}
\end{split}
$$

For additional data for estimation the electrode arrangement can be used in bipolar mode when the voltage is measured from the current injecting electrodes. The bipolar mode will give 4 distinct impedance measurements:
1. between electrodes 1 and 4
2. between electrodes 2 and 3
3. between electrodes 1 and 2
4. between electrodes 1 and 3
