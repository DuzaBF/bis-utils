# Multiplexing 4 electrode array for bioimpedance spectroscopy of the limbs
Using two-layer half-space model with am assumption of the quasi static case.

Electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma_1} \cdot \left[\frac{1}{\sqrt{r^2 + z^2}} + \sum^{\infty}_{n=1}K^{n}_{12}\left(\frac{1}{\sqrt{r^2 + (2nd_1-z)^2}} + \frac{1}{\sqrt{r^2 + (2nd_1+z)^2}} \right) \right],
$$
where 

$I$ - injected current;

$K_{12} = \frac{\sigma_1 - \sigma_2}{\sigma_1 + \sigma_2}$ - reflection coefficient; 

$\sigma_1$ - conductivity of the first layer;

$\sigma_2$ - conductivity of the second layer;

$d_1$ - thickness of the first layer;

Electric field measured on the surfce is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma_1} \cdot \left[\frac{1}{r} + 2\sum^{\infty}_{n=1}\frac{K^{n}_{12}}{\sqrt{r^2 + (2nd_1)^2}} \right],
$$
