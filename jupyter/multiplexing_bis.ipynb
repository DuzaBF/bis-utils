# Multiplexing 4 electrode array for bioimpedance spectroscopy of the limbs
Using one-layer and two-layer half-space model with am assumption of the quasi static case.

# One-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{1}{\sqrt{r^2 + z^2}},
$$
where 

$I$ - injected current;

$\sigma$ - conductivity of the half-space material;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{r}{(r^2 + z^2)^{3/2}}
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{z}{(r^2 + z^2)^{3/2}}
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma} \cdot \frac{1}{r}
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma} \cdot \frac{1}{r^2}
$$
$$
E_z = 0
$$

# Two-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{\sqrt{r^2 + z^2}} + 
    \sum^{\infty}_{n=1}K^{n}_{12}
    \left(
        \frac{1}{\sqrt{r^2 + (2nd_1-z)^2}} + 
        \frac{1}{\sqrt{r^2 + (2nd_1+z)^2}} 
    \right) 
\right],
$$
where 

$I$ - injected current;

$K_{12} = \frac{\sigma_1 - \sigma_2}{\sigma_1 + \sigma_2}$ - reflection coefficient; 

$\sigma_1$ - conductivity of the first layer;

$\sigma_2$ - conductivity of the second layer;

$d_1$ - thickness of the first layer;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{r}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{rK^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{z}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{-(2nd_1 - z)K^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{(2nd_1 + z)K^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{r} + 
    2\sum^{\infty}_{n=1}\frac{K^{n}_{12}}{\sqrt{r^2 + (2nd_1)^2}} 
\right],
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{r^2} + 
    2\sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1)^2)^{3/2}} 
    \right) 
\right]
$$
$$
E_z = 0
$$
