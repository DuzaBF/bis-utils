# Multiplexing 4 electrode array for bioimpedance spectroscopy of the limbs
Using one-layer and two-layer half-space model with am assumption of the quasi static case.

# One-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{1}{\sqrt{r^2 + z^2}},
$$
where 

$I$ - injected current;

$\sigma$ - conductivity of the half-space material;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{r}{(r^2 + z^2)^{3/2}}
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma} \cdot 
    \frac{z}{(r^2 + z^2)^{3/2}}
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma} \cdot \frac{1}{r}
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma} \cdot \frac{1}{r^2}
$$
$$
E_z = 0
$$

# Two-layer model
Potential of the electric field generated by the point current source placed on the surface of the object:
$$
V(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{\sqrt{r^2 + z^2}} + 
    \sum^{\infty}_{n=1}K^{n}_{12}
    \left(
        \frac{1}{\sqrt{r^2 + (2nd_1-z)^2}} + 
        \frac{1}{\sqrt{r^2 + (2nd_1+z)^2}} 
    \right) 
\right],
$$
where 

$I$ - injected current;

$K_{12} = \frac{\sigma_1 - \sigma_2}{\sigma_1 + \sigma_2}$ - reflection coefficient; 

$\sigma_1$ - conductivity of the first layer;

$\sigma_2$ - conductivity of the second layer;

$d_1$ - thickness of the first layer;

The electrical field strength $E$ is defined as:
$$
\vec{E}(r, z) = 
-\nabla V = 
-\frac{\partial V}{\partial r} \vec{r} - \frac{\partial V}{\partial z} \vec{z} 
$$

$$
E_r(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{r}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{rK^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

$$
E_z(r, z) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{z}{(r^2 + z^2)^{3/2}} + 
    \sum^{\infty}_{n=1}
    \left(
            \frac{-(2nd_1 - z)K^{n}_{12}}{(r^2 + (2nd_1-z)^2)^{3/2}} + \frac{(2nd_1 + z)K^{n}_{12}}{(r^2 + (2nd_1+z)^2)^{3/2}} 
    \right) 
\right]
$$

Potential of the electric field measured on the surface is determined by setting $z = 0$:
$$
V(r) = \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{r} + 
    2\sum^{\infty}_{n=1}\frac{K^{n}_{12}}{\sqrt{r^2 + (2nd_1)^2}} 
\right],
$$

Electric field strength on the surface:
$$
E_r(r)= \frac{I}{2\pi\sigma_1} \cdot 
\left[
    \frac{1}{r^2} + 
    2\sum^{\infty}_{n=1}
    \left(
            \frac{rK^{n}_{12}}{(r^2 + (2nd_1)^2)^{3/2}} 
    \right) 
\right]
$$
$$
E_z = 0
$$

# Tetrapolar electrode system
In symmetrical tetrapolar electrode system the current is usually injected through the electrodes denoted A and B and voltage is measured between electrodes denoted M and N. The coordinates of the electrodes are denoted as $x_A$, $x_B$, $x_M$, $x_N$. The distance between electrodes A and B is $L$, between electrodes M and N is $s$. On the surface the potential will depend only on the $r=x - x_A$.
$$
AB = x_B - x_A = L 
$$
$$
MN = x_N - x_M = s 
$$
$$
AM = BN = x_M - x_A = x_N - x_B = \frac{L-s}{2} 
$$
$$
AN = BM = x_N - x_A = x_M - x_B = \frac{L+s}{2} 
$$

All electrodes are the same point sources, so the potential from the electrode depends only on the distance to the measuring point. 

For simplicity:
$$
q=\frac{I}{2 \pi \sigma_1}
$$
The potentials at points M is determined as a sum of potentials from electrode A with current $I$ and electrode B with current $-I$:
$$
V_M = V(x_M - x_A) + V(x_M - x_B) = q \cdot 
\left[
        \frac{1}{x_M - x_A} - \frac{1}{x_M - x_B} + 
        2\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} 
        \right)
\right]
$$
$$
V_N = V(x_N - x_A) + V(x_N - x_B) = q \cdot 
\left[
        \frac{1}{x_N - x_A} - \frac{1}{x_N - x_B} +
        2\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}} 
        \right)
\right]
$$

Voltage between electrodes:
$$
\begin{split}
U_{MN} = V_M - V_N  = q \cdot
&\left[
        \frac{1}{x_M - x_A} - \frac{1}{x_M - x_B} + \frac{1}{x_N - x_B} - \frac{1}{x_N - x_A} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_M - x_A)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_A)^2 + (2nd_1)^2}} 
        \right)
\right] = \\
= 
&\left[
        \frac{4}{L-s} - \frac{4}{L+s}
        + 4\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(\frac{L-s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}}
        \right)
\right]
\end{split}
$$

Measured impedance:
$$
Z_{MN} = \frac{U_{MN}}{I} = \frac{1}{2 \pi \sigma_1} \cdot 
\left[
        \frac{4}{L-s} - \frac{4}{L+s} + 
        4\sum^{\infty}_{n=1}
        K^{n}_{12}
        \left(
            \frac{1}{\sqrt{(\frac{L-s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} 
        \right)
\right]
$$

By changing which pairs of the electrodes are current injecting and which are voltage measuring different impedance will be measured. Due to symmetry and reciprocity only 3 arrangements are distinct:
1. AB - current, MN - measuring;
2. BN - current, AM - measuring;
3. BM - current, AN - measuring.

The measured impedances in case 2 are determined by:
$$
\begin{split}
Z_{AM} = \frac{1}{2 \pi \sigma_1} \cdot
&\left[
        \frac{1}{x_A - x_B} - \frac{1}{x_A - x_N} + \frac{1}{x_M - x_B} - \frac{1}{x_M - x_N} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_A - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_A - x_N)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_M - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_M - x_N)^2 + (2nd_1)^2}} 
        \right)
\right] = \\
= 
&\left[
        \frac{1}{L} - \frac{1}{\frac{L+s}{2}} + \frac{1}{\frac{L+s}{2}} - \frac{1}{s} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(L)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(s)^2 + (2nd_1)^2}} 
        \right)
\right]
\end{split}
$$
$$
Z_{AM} = \frac{1}{2 \pi \sigma_1} \cdot
\left[
        \frac{1}{L} - \frac{1}{s}
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}.\left(
            \frac{1}{\sqrt{L^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{s^2 + (2nd_1)^2}} 
        \right)
\right]
$$
The measured impedances in case 3 are determined by:
$$
\begin{split}
Z_{AN} = \frac{1}{2 \pi \sigma_1} \cdot
&\left[
        \frac{1}{x_A - x_B} - \frac{1}{x_A - x_M} + \frac{1}{x_N - x_B} - \frac{1}{x_N - x_M} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(x_A - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_A - x_M)^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(x_N - x_B)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(x_N - x_M)^2 + (2nd_1)^2}} 
        \right)
\right] = \\
= 
&\left[
        \frac{1}{L} - \frac{1}{\frac{L-s}{2}} + \frac{1}{\frac{L-s}{2}} - \frac{1}{s} + \right.\\
        \left.
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}\right.
        &\left.\left(
            \frac{1}{\sqrt{(L)^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} +
            \frac{1}{\sqrt{(\frac{L+s}{2})^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{(s)^2 + (2nd_1)^2}} 
        \right)
\right]
\end{split}
$$
$$
Z_{AN} = \frac{1}{2 \pi \sigma_1} \cdot
\left[
        \frac{1}{L} - \frac{1}{s}
        + 2\sum^{\infty}_{n=1}
        K^{n}_{12}.\left(
            \frac{1}{\sqrt{L^2 + (2nd_1)^2}} -
            \frac{1}{\sqrt{s^2 + (2nd_1)^2}} 
        \right)
\right]
$$
